```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.fft import fft2, ifft2, fftshift
from PIL import Image
```

```python
citra = np.array(Image.open('../images/cruise-periodicNoise.jpg').convert('L'))  # Ubah 'L' menjadi 'RGB' jika citra berwarna
```

```python
Ifreq = fft2(citra)
Ifreq = fftshift(Ifreq)
```

```python
Ifreq[:, 158] = 0
Ifreq[:, 194] = 0
Ifreq[165, :] = 0
Ifreq[187, :] = 0

Ifreq_inv = ifft2(Ifreq)
```

```python
plt.figure(figsize=(10, 8))
plt.subplot(2, 2, 1)
plt.imshow(citra, cmap='gray')
plt.title('Original Image')
plt.axis('off')

plt.subplot(2, 2, 2)
plt.imshow(np.log(np.abs(Ifreq) + 1), cmap='gray')
plt.title('Frequency Spectrum')
plt.axis('off')

plt.subplot(2, 2, 3)
plt.imshow(np.abs(Ifreq_inv), cmap='gray')
plt.title('Processed Image')
plt.axis('off')

plt.subplot(2, 2, 4)
plt.imshow(np.log(np.abs(Ifreq) + 1), cmap='gray')
plt.title('Filtered Frequency Spectrum')
plt.axis('off')

plt.tight_layout()
plt.show()
```
